#### Get a sentiment score for the text
Unfortunately SpaCy doesn't have sentiment scores built in (there is an attribute for it, but it only works on specific models).
However TextBlob has us covered with the `.sentiment` attribute. Just wrap a string in a `TextBlob` wrapper and you're ready to go.

`.sentiment` gives us access to two scores
- *Polarity* ranging from -1 to 1, meaning very negative to very positive in sentiment. 0 is neutral.
- *Subjectivity* ranging from 0 to 1, 0 is absolutely objective, 1 is absolutely subjective.

Both these scores are derived from rules around the words used within a piece of text, but language is often tricky. These scores are useful but should be treated with well cultivated skepticism.

```python
happy = "Life is just fantastic now that I have found love and joy!"
sad = "I am a worthless husk after losing everything and I hate everyone"

happy_doc = textblob.TextBlob(happy)
print(happy_doc)
print(happy_doc.sentiment)

print()

sad_doc = textblob.TextBlob(sad)
print(sad_doc)
print(sad_doc.sentiment)
```

```python
# You can access the polarity and the subjectivity seperately using the associated .polarity and .subjectivity attributes
print(happy_doc.sentiment.polarity)
print(happy_doc.sentiment.subjectivity)
```
