```python
from spotipy.oauth2 import SpotifyClientCredentials
import spotipy
import pandas as pd


client_credentials_manager = SpotifyClientCredentials(client_id='YOUR_CLIENT_ID_HERE',
                                                      client_secret='YOUR_SECRET_HERE')
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)

uri = 'spotify:user:spotifycharts:playlist:37i9dQZEVXbLnolsZ8PSNw' # URI for the Spotify UK top 50 playlist.
username = uri.split(':')[2]
playlist_id = uri.split(':')[4]

results = sp.user_playlist(username, playlist_id)
```

```python
df = pd.DataFrame(results['tracks']['items'])[['track']]
```

```python
def get_genres(artists_uri):
    try:
        return sp.artist(artists_uri)['genres'][0][0]
    except IndexError:
        return pd.np.NAN
        
```

```python
new_df = pd.DataFrame()
new_df['track_id'] = df.track.apply(lambda x: x['id'])
new_df['track_name'] = df.track.apply(lambda x: x['name'])
new_df['artists'] = df.track.apply(lambda x: [y['name'] for y in x['artists']])
new_df['genre'] = df.track.apply(lambda x: get_genres(artists_uri=x['artists'][0]['uri']))
new_df['release_date'] = df.track.apply(lambda x: x['album']['release_date'])
new_df['explicit'] = df.track.apply(lambda x: x['explicit'])
new_df['rank'] = df.index +1
new_df['popularity'] = df.track.apply(lambda x: x['popularity'])
new_df['duration_ms'] = df.track.apply(lambda x: x['duration_ms'])
```

```python
audio = pd.DataFrame(sp.audio_features(new_df.track_id))
```

```python
keep_cols = ['id','danceability','loudness','speechiness']

new_df = new_df.merge(audio[keep_cols], left_on='track_id', right_on='id')
del new_df['id']
new_df
```

```python
new_df.to_csv('spotify_top_50.csv')
```
